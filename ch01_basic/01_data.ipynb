```python
X = torch.tensor([[1.0,2.0],[3.0,4.0]])
Y = torch.tensor([[0.1,0.2],[0.3,0.4]])
torch.dot(X.reshape(-1),Y.reshape(-1))#点积，对应元素相乘相加
x = torch.tensor([2,1],dtype=torch.float32)
torch.mv(X,x)#矩阵与向量乘法
torch.mm(X,Y)#矩阵乘法
torch.norm(X)
```

1.3 微分

```python
x = torch.tensor([1.0,2.0,3.0],dtype=torch.float32,requires_grad=True)
y = 2 * torch.dot(x,x)
y.backward()
x.grad.zero_()#梯度清零
y = x.sum()
y.backward()
x.grad.zero_()
y = x * x
y.sum().backward()
x.grad

```

```python
x = torch.tensor([5.0],dtype=torch.float32,requires_grad=True)
y = x ** 2
y.backward()
y,x.grad
```
